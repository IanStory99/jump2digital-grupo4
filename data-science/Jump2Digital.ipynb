# Addirional features and global dataframe

```python
df_products ['cat_cuisine'] = df_products ['category']+'_'+df_products ['cuisine'] + '_' + [str(x) for x in df_products ['meal_id']]
```

```python
df_products
```

```python
df_demand['dif_price'] = df_demand['checkout_price'] - df_demand['base_price'] 
```

```python
df_demand['billed_money'] = df_demand['checkout_price'] * df_demand['num_orders']
```

```python
demand_products = pd.merge(df_demand, df_products, on = "meal_id", how = "left")
```

```python

demand_products_centers = pd.merge (demand_products, df_centers, on = "center_id", how = "left")
```

```python
demand_products_centers
```

```python
top_10_products = list(demand_products_centers.groupby('meal_id').sum().nlargest(10, 'billed_money').sort_values('billed_money', ascending = False).index)
top_10_centers =  list(demand_products_centers.groupby('center_id').sum().nlargest(10, 'billed_money').sort_values('billed_money', ascending = False).index)

top_10_cities = list(demand_products_centers.groupby('city_code').sum().nlargest(10, 'billed_money').sort_values('billed_money', ascending = False).index)


top_3_areas = list(demand_products_centers.groupby('region_code').sum().nlargest(3, 'billed_money').sort_values('billed_money', ascending = False).index)
```

```python

list_centers =list (demand_products_centers.center_id.unique())
list_products = list (demand_products_centers.meal_id.unique())

```
